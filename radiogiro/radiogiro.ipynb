```python
import matplotlib.pyplot as plt

# Funcin para leer los datos de los archivos .xvg
def read_xvg(filename):
    time = []
    gyration_radius = []
    with open(filename, 'r') as file:
        for line in file:
            if not line.startswith('#') and not line.startswith('@'):
                columns = line.split()
                time.append(float(columns[0]))  # Tiempo en el primer columna (en ps)
                gyration_radius.append(float(columns[1]))  # Radio de giro en la segunda columna (en nm)
    return time, gyration_radius

# Leer los datos de ambos archivos
time_298K, gyration_298K = read_xvg('gyrate_298K.xvg')
time_400K, gyration_400K = read_xvg('gyrate_400K.xvg')

# Graficar los resultados
plt.figure(figsize=(10, 6))

# Graficar para 298K
plt.plot(time_298K, gyration_298K, label='298K', color='b', linewidth=2)

# Graficar para 400K
plt.plot(time_400K, gyration_400K, label='400K', color='r', linewidth=2)

# Etiquetas y ttulo
plt.xlabel('Tiempo (ps)')
plt.ylabel('Radio de Giro (nm)')
plt.title('Evolucin del Radio de Giro a 298K y 400K')

# Leyenda
plt.legend()

# Mostrar el grfico
plt.grid(True)
plt.tight_layout()

# Guardar la imagen en formato PNG
plt.savefig('gyration_radius_comparison.png', dpi=300)

# Mostrar el grfico
plt.show()



```
