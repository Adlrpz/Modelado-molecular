```python
import numpy as np
import matplotlib.pyplot as plt

# Función para leer archivos .xvg de GROMACS
def leer_xvg(filename):
    data = []
    with open(filename, 'r') as f:
        for line in f:
            if not line.startswith(('#', '@')):  # Ignorar comentarios y encabezados
                data.append([float(x) for x in line.split()])
    return np.array(data)

# Cargar los datos de las simulaciones a 298 K y 400 K
data_298K = leer_xvg("/home/alumno06/PMM/graficas/energiatotal/298/energy.xvg")
data_400K = leer_xvg("/home/alumno06/PMM/graficas/energiatotal/400/energy.xvg")

# Extraer el tiempo y la energía total
time_298, energy_298 = data_298K[:,0], data_298K[:,1]
time_400, energy_400 = data_400K[:,0], data_400K[:,1]

# Graficar Energía Total a 298 K
plt.figure(figsize=(8, 5))
plt.plot(time_298, energy_298, label="298 K", color='blue')
plt.xlabel("Tiempo (ps)")
plt.ylabel("Energía Total (kJ/mol)")
plt.title("Energía Total del Sistema a 298 K")
plt.legend()
plt.grid()
plt.savefig("energia_total_298K.jpg", dpi=300)  # Guardar la imagen
plt.show()

# Graficar Energía Total a 400 K
plt.figure(figsize=(8, 5))
plt.plot(time_400, energy_400, label="400 K", color='red')
plt.xlabel("Tiempo (ps)")
plt.ylabel("Energía Total (kJ/mol)")
plt.title("Energía Total del Sistema a 400 K")
plt.legend()
plt.grid()
plt.savefig("energia_total_400K.jpg", dpi=300)  # Guardar la imagen
plt.show()

# Graficar la energía total
plt.figure(figsize=(8,5))
plt.plot(time_298, energy_298, label="298 K", color="blue")
plt.plot(time_400, energy_400, label="400 K", color="red")
plt.xlabel("Tiempo (ps)")
plt.ylabel("Energía Total (kJ/mol)")
plt.title("Evolución de la Energía Total en la Simulación")
plt.legend()
plt.grid()
plt.savefig("energia_total_conjunto.jpg", dpi=300)  # Guardar la imagen
plt.show()
```
