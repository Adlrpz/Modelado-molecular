```python
import numpy as np
import matplotlib.pyplot as plt

# Cargar datos
phi_data = np.loadtxt("phi-arg-3.dat", usecols=[1])
psi_data = np.loadtxt("psi-arg-3.dat", usecols=[1])

# Crear un eje de tiempo basado en el nmero de datos
time_steps = np.arange(len(phi_data))

# Grfica de evolucin de 
plt.figure(figsize=(8, 5))
plt.plot(time_steps, phi_data, marker="o", linestyle="-", markersize=2, label="ngulo ")
plt.xlabel("Paso de tiempo")
plt.ylabel("ngulo  (grados)")
plt.title("Evolucin del ngulo  de ARG-3")
plt.grid()
plt.legend()
plt.savefig("phi_evolucion.png", dpi=300)  # Guardar figura
plt.show()

# Grfica de evolucin de 
plt.figure(figsize=(8, 5))
plt.plot(time_steps, psi_data, marker="o", linestyle="-", markersize=2, label="ngulo ", color="red")
plt.xlabel("Paso de tiempo (fs)")
plt.ylabel("ngulo  (grados)")
plt.title("Evolucin del ngulo  de ARG-3")
plt.grid()
plt.legend()
plt.savefig("psi_evolucion.png", dpi=300)
plt.show()

# Mapa de Ramachandran ( vs )
plt.figure(figsize=(7, 7))
plt.scatter(phi_data, psi_data, c="blue", alpha=0.5, s=5)  # Scatter plot
plt.xlabel("ngulo  (grados)")
plt.ylabel("ngulo  (grados)")
plt.title("Mapa de Ramachandran de ARG-3")
plt.xlim(25, 80)
plt.ylim(25, 80)
plt.grid()
plt.savefig("ramachandran.png", dpi=300)
plt.show()
```
