```python
import numpy as np
import matplotlib.pyplot as plt

# Funcin para leer archivos .xvg de GROMACS
def leer_xvg(filename):
    data = []
    with open(filename, 'r') as f:
        for line in f:
            if not line.startswith(('#', '@')):  # Ignorar comentarios y encabezados
                data.append([float(x) for x in line.split()])
    return np.array(data)

# Cargar los datos de las simulaciones a 298 K y 400 K
data_298K = leer_xvg("/home/alumno06/PMM/graficas/energiacine//298/energy.xvg")
data_400K = leer_xvg("/home/alumno06/PMM/graficas/energiacine/400/energy.xvg")

# Extraer el tiempo y la energa total
time_298, energy_298 = data_298K[:,0], data_298K[:,1]
time_400, energy_400 = data_400K[:,0], data_400K[:,1]

# Graficar Energa Total a 298 K
plt.figure(figsize=(8, 5))
plt.plot(time_298, energy_298, label="298 K", color='blue')
plt.xlabel("Tiempo (ps)")
plt.ylabel("Energa cintica (kJ/mol)")
plt.title("Energa cintica del sistema a 298 K")
plt.legend()
plt.grid()
plt.savefig("energia_totalc_298K.jpg", dpi=300)  # Guardar la imagen
plt.show()

# Graficar Energa Total a 400 K
plt.figure(figsize=(8, 5))
plt.plot(time_400, energy_400, label="400 K", color='red')
plt.xlabel("Tiempo (ps)")
plt.ylabel("Energa cintica (kJ/mol)")
plt.title("Energa cintica del sistema a 400 K")
plt.legend()
plt.grid()
plt.savefig("energia_totalc_400K.jpg", dpi=300)  # Guardar la imagen
plt.show()

# Graficar la energa total
plt.figure(figsize=(8,5))
plt.plot(time_298, energy_298, label="298 K", color="blue")
plt.plot(time_400, energy_400, label="400 K", color="red")
plt.xlabel("Tiempo (ps)")
plt.ylabel("Energa cintica (kJ/mol)")
plt.title("Evolucin de la energa cintica en la simulacin")
plt.legend()
plt.grid()
plt.savefig("energia_totalc_conjunto.jpg", dpi=300)  # Guardar la imagen
plt.show()
```
