```python
import matplotlib.pyplot as plt
import numpy as np

# Rutas a los archivos de datos (ajusta estas rutas seg√∫n tu estructura de carpetas)
file_298K = '298/angaver.xvg'  # Datos a 298 K
file_400K = '400/angaver.xvg'  # Datos a 400 K

# Funci√≥n para leer los datos del archivo .xvg
def read_xvg(file_path):
    time = []
    angle = []
    with open(file_path, 'r') as file:
        for line in file:
            if not line.startswith(('#', '@')):  # Ignorar l√≠neas de comentarios
                data = line.split()
                time.append(float(data[0]))  # Primera columna: tiempo
                angle.append(float(data[1]))  # Segunda columna: √°ngulo
    return np.array(time), np.array(angle)

# Leer los datos para ambas temperaturas
time_298K, angle_298K = read_xvg(file_298K)
time_400K, angle_400K = read_xvg(file_400K)

# Funci√≥n para crear y guardar una gr√°fica
def plot_angle_vs_time(time, angle, temperature, color, output_file):
    plt.figure(figsize=(10, 6))
    plt.plot(time, angle, label=f'{temperature} K', color=color, linestyle='-', linewidth=1.5, marker='o', markersize=4)
    plt.title(f'Variaci√≥n del √°ngulo en funci√≥n del tiempo a {temperature} K', fontsize=16)
    plt.xlabel('Tiempo (ps)', fontsize=14)
    plt.ylabel('√Ångulo (¬∞)', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend(fontsize=12)
    plt.tick_params(axis='both', which='major', labelsize=12)
    plt.xlim(left=0)
    plt.ylim(bottom=min(angle) - 5, top=max(angle) + 5)
    plt.tight_layout()
    plt.savefig(output_file, dpi=300, bbox_inches='tight')  # Guardar en alta resoluci√≥n (300 dpi)
    print(f"Gr√°fica guardada como {output_file}")
    plt.show()

# Crear y guardar la gr√°fica para 298 K
plot_angle_vs_time(time_298K, angle_298K, 298, 'blue', 'angulo_vs_tiempo_298K.png')

# Crear y guardar la gr√°fica para 400 K
plot_angle_vs_time(time_400K, angle_400K, 400, 'red', 'angulo_vs_tiempo_400K.png')
```
